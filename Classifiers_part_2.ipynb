## Helpers

```python
all_scores = pd.DataFrame(columns=["Classifier", "Precision", "Recall", "F-Score", "ROC_AUC", "Accuracy"])
y_predictions = dict()
```

```python
def scores(y_true, y_pred, y_pred_prob):
    prec, recall, f1, n = p_r_f1_scores(y_true, y_pred, average='binary')
    roc_auc = roc_auc_score(y_true, y_pred_prob)
    acc = accuracy_score(y_true, y_pred)
    return [prec, recall, f1, roc_auc, acc]
```

```python
def metrics(y_true, y_pred, y_pred_prob_posi): 
    print("\nEvaluation Metrics\n")
    
    # Confusion Matrix where C[i][j] represents true value is in class i and predicted to be in class j
    # C[1][1] true value is class 1 and predicted to be in class 1 
    # That is C[1][1] is True Positive
    # C[0][0] is True Negative
    cm = confusion_matrix(y_true, y_pred)
    cm_matrix = pd.DataFrame(data=cm, index=["Actual Negative: 0", "Actual Positive: 1"], 
                             columns=["Predicted Negative: 0", "Predicted Postive: 1"])

    sns.heatmap(cm_matrix, annot=True, fmt='d')

    print(classification_report(y_true, y_pred))
    # TPR = TP / (TP + FN)
    # FPR = FP / (FP + TN)
    fpr, tpr, thres = roc_curve(y_true, y_pred_prob_posi)
    plt.figure()
    plt.plot(fpr, tpr, linewidth="2")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("False positive rate")
    plt.ylabel("True positive rate")
    plt.show()
    
    auc = roc_auc_score(y_true, y_pred_prob_posi)
    print("roc_auc: ", auc)
```

##### Some of them occur only once 

percentage of frequency distibution of countries

```python
print(df["country"].value_counts() / df.shape[0])
```

We will try to decrease number of classes of 'country' by assigning a class for each country that has frequency above 0.1 and the others will be grouped in one group that be considered as 'other' or 'else' group
(Binning)

```python
newCountry = pd.DataFrame(data=df['country'], columns=["country"]) 
# print(newCountry.head())
newCounter = 1
for country, freq in \
        zip(df['country'].value_counts().keys().to_list(), df['country'].value_counts().to_list()):
    if freq / df.shape[0] >= 0.1:
        newCountry.loc[df['country'] == country, 'country'] = newCounter
        newCounter += 1
    else:
        newCountry.loc[df['country'] == country, 'country'] = 0
```

```python
print(newCountry.value_counts())
```

Now it has 4 unique variables 1,2,3 for countries and '0' represents other countries

Now we will one hot encode country

```python
newCountry = pd.get_dummies(newCountry, columns=['country'])
```

```python
newCountry.shape
```

```python
df = df.drop(['country'], axis=1)
df = pd.concat([df, newCountry], axis=1)
```

### We will look at symptoms

```python
symptoms = ["symptom1", "symptom2", "symptom3", "symptom4", "symptom5", "symptom6"]
for sym in symptoms:
    print(df[sym].value_counts())
```

In Symptoms 1 to 4

```python
df[["symptom1", "symptom2", 'symptom3', 'symptom4']]
```

##### Symptom1 has alot of values that doesn;t have high frequency 

##### I will take a similar approach to what have been done to 'country' feature above

```python
changedSymp = ['symptom1', 'symptom2', 'symptom3', 'symptom4']
newSymptoms = pd.DataFrame(data=df[changedSymp])
for symptom in changedSymp:
    counter = 1
    for symi, cnt in \
        zip(df[symptom].value_counts().keys().to_list(), df[symptom].value_counts().to_list()):
        if cnt / df.shape[0] >= 0.05:
            newSymptoms.loc[df[symptom] == symi, symptom] = counter
            counter += 1
        else:
            newSymptoms.loc[df[symptom] == symi, symptom] = 0
```

```python
for i in changedSymp:
    print(newSymptoms[i].value_counts())
```

```python
df = df.drop(changedSymp, axis=1)
df = pd.concat([df, newSymptoms[changedSymp]], axis=1)
```

```python
df
```

### We can see that the target variable is imbalanced

```python
df["result"].value_counts() / len(df)
```

Since the target class is imbalanced we will keep an eye on metrics other than accuracy as F1, Recall and Precision

### Spliting features from target

We will drop location since it suffers from high cardinality and part of the information that location feature gives is already related to country feature as location is part of the country

```python
y = df["result"]
X = df.drop(["result", "location"], axis=1)
print("y dimensions:{} X dimesions:{}".format(y.shape,X.shape))
```

### Spliting data to training and test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, \
                                                    random_state=22, shuffle=True, stratify=y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
print(y_train.value_counts(), '\n',y_test.value_counts())
```

```python
X_train.head(5)
```



# Decision Trees



```python
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)
```

```python
tree_clf_cv = cross_val_score(tree_clf, X_train, y_train, 
                              cv=StratifiedKFold(n_splits=5, random_state=26, shuffle=True), 
                              scoring='recall', n_jobs=-1)
print("cross validation scores: {}, and their mean: {}".format(tree_clf_cv, np.mean(tree_clf_cv)))
```

### Prediciting on test data set

```python
tree_clf_test_predict = tree_clf.predict(X_test)
tree_clf_test_predict_proba = tree_clf.predict_proba(X_test)[:, 1]
metrics(y_test, tree_clf_test_predict, tree_clf_test_predict_proba)
```

Comparing test data metrics to training data set metrics below

```python
tree_clf_train_predict = tree_clf.predict(X_train)
p_r_f1_scores(y_train, tree_clf_train_predict, average='binary')[1]
```

Recall on training data set is about 99% and for testing data set is 91% so probably this classifier is overfitting

### Tuning Hyperparameters


```python
str_kf = StratifiedKFold(n_splits=5, random_state=26, shuffle=True)
```

```python
parameters = {
    'criterion': ['entropy', 'gini'], 
    'max_depth': np.arange(2, 16, ),
    'min_samples_leaf': np.arange(2, 20, 2),
    'min_samples_split': np.arange(2, 15, 2),
    'class_weight': [{0:1, 1:1.2}],
}
```

```python
gcv_tree = GridSearchCV(estimator=DecisionTreeClassifier(random_state=42), 
                        param_grid=parameters, scoring='recall', refit=True, n_jobs=-1, 
                        cv=str_kf)
gcv_tree.fit(X_train, y_train)
```

```python
print("Best tree: {} \nBest Score: {}".format(gcv_tree.best_params_, gcv_tree.best_score_))
```

mean of cross validation scores increased

## Prediciting on Test data set

```python
gcv_test_predict = gcv_tree.predict(X_test)
gcv_test_predict_proba = gcv_tree.predict_proba(X_test)[:, 1]
metrics(y_test, gcv_test_predict, gcv_test_predict_proba)
```

roc_auc value increased too

Recall value on testing data set is close to recall value on training data set 

```python
gcv_train_predict = gcv_tree.predict(X_train)
p_r_f1_scores(y_train, gcv_train_predict, average='binary')[1]
```

```python
sc = scores(y_test, gcv_test_predict, gcv_test_predict_proba)
all_scores.loc["tree"] = ["Decision tree"] + sc
y_predictions["Decision tree"] = [gcv_test_predict, gcv_test_predict_proba]
```

## Visualizing tree

# SVM

```python
svc = SVC(probability=True, random_state=42)
svc.fit(X_train, y_train)
```

```python
svc_test_predict = svc.predict(X_test)
svc_test_predict_proba = svc.predict_proba(X_test)[:, 1]
metrics(y_test, svc_test_predict, svc_test_predict_proba)
```

### Tuning Hyperparameters

```python
svc_parameters = [
    {'C': [0.01, 0.1, 1, 10, 100], 'kernel': ['linear']},
    {'C': [0.01, 0.1, 1, 10, 100], 'kernel': ['rbf'], 'gamma': [100, 10, 1, 0.1, 0.01, 0.001]},
    {'C': [0.01, 0.1, 1, 10, 100], 'kernel': ['poly'], 'degree': [2, 3, 4], }
]
```

```python
svc_grid = GridSearchCV(estimator=SVC(class_weight={0: 1, 1: 2}, probability=True, random_state=22), 
                        param_grid=svc_parameters, 
                        cv=StratifiedKFold(n_splits=3, shuffle=True, random_state=22), 
                        n_jobs=-1, refit=True, scoring='recall', verbose=1)
svc_grid.fit(X_train, y_train)
```

```python
print('Best parameters: {}, Best score: {}'.format(svc_grid.best_params_, svc_grid.best_score_))
```

```python
svc_g_test_predict = svc_grid.predict(X_test)
svc_g_test_predict_proba = svc_grid.predict_proba(X_test)[:, 1]
metrics(y_test, svc_g_test_predict, svc_g_test_predict_proba)
```

Overall classifier got better than the default one

```python
all_scores.loc["svc"] = ["Support Vector Machine"] + scores(y_test, svc_g_test_predict, svc_g_test_predict_proba)
y_predictions["Support Vector Machine"] = [svc_g_test_predict, svc_g_test_predict_proba]
```
